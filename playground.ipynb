helper method which takes in the replay name,boolean, and variable and calls the bash command on those things.
It returns the "list" which should be changed into the correct datastructure.

```python
def ripData(replayName, jsonBool, thingy):
    thing = "python heroprotocol.py "
    if(jsonBool):
        thing = thing + "--json "
    thing = thing + thingy + " \"" + replayName+ "\" "
    #print(thing)
    tempThing = ! {thing}
    return tempThing
    
```
